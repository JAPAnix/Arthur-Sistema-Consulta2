## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**

- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2
# Adicionando m√©todos de a√ß√£o na classe principal

class Consulta_Arthur:

    def __init__(self, paciente, medico, data):
        self.paciente = paciente
        self.medico = medico
        self.data = data

    def exibir_dados_Arthur(self):
        print(f"Paciente: {self.paciente}")
        print(f"M√©dico: {self.medico}")
        print(f"Data: {self.data}")

    # M√©todo para alterar a data da consulta
    def alterar_data_Arthur(self, nova_data):
        self.data = nova_data
        print(f"Data alterada para: {self.data}")

    # M√©todo para alterar o m√©dico
    def alterar_medico_Arthur(self, novo_medico):
        self.medico = novo_medico
        print(f"M√©dico alterado para: {self.medico}")

# Subclasse: ConsultaEspecializada_Arthur

class ConsultaEspecializada_Arthur(Consulta_Arthur):

    def __init__(self, paciente, medico, data, especialidade):
        super().__init__(paciente, medico, data)  # Chamando o construtor da superclasse
        self.especialidade = especialidade  # Novo atributo exclusivo da subclasse

    # Novo m√©todo exclusivo da subclasse
    def exibir_especialidade_Arthur(self):
        print(f"Especialidade: {self.especialidade}")

# Criando objeto da subclasse
consulta_especial = ConsultaEspecializada_Arthur("Pedro Lima", "Dra. Beatriz", "2025-06-20", "Cardiologia")

# Exibindo dados e especialidade
consulta_especial.exibir_dados_Arthur()
consulta_especial.exibir_especialidade_Arthur()
```

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 5
import mysql.connector
from datetime import datetime

# Classe principal
class Consulta_Arthur:

    def __init__(self, paciente, medico, data):
        self.__paciente = paciente
        self.medico = medico
        self.data = data

    # Getter e Setter para paciente
    def get_paciente_Arthur(self):
        return self.__paciente

    def set_paciente_Arthur(self, novo_paciente):
        self.__paciente = novo_paciente

    # Exibir dados
    def exibir_dados_Arthur(self):
        print(f"Paciente: {self.__paciente}")
        print(f"M√©dico: {self.medico}")
        print(f"Data: {self.data}")

    # Verificar se a data √© futura
    def verificar_data_Arthur(self):
        data_consulta = datetime.strptime(self.data, "%Y-%m-%d")
        hoje = datetime.today()
        if data_consulta > hoje:
            print("A consulta est√° agendada para o futuro.")
        else:
            print("A data da consulta j√° passou ou √© hoje.")

    # Salvar no MySQL
    def salvar_no_banco_Arthur(self, conexao):
        cursor = conexao.cursor()
        sql = "INSERT INTO consultas (paciente, medico, data) VALUES (%s, %s, %s)"
        valores = (self.__paciente, self.medico, self.data)
        cursor.execute(sql, valores)
        conexao.commit()
        print("Consulta salva no banco de dados MySQL.")

# Subclasse com polimorfismo
class ConsultaEspecializada_Arthur(Consulta_Arthur):

    def __init__(self, paciente, medico, data, especialidade):
        super().__init__(paciente, medico, data)
        self.especialidade = especialidade

    # Sobrescrevendo o m√©todo de exibi√ß√£o
    def exibir_dados_Arthur(self):
        print(f"Paciente: {self.get_paciente_Arthur()}")
        print(f"M√©dico: {self.medico} - Especialidade: {self.especialidade}")
        print(f"Data: {self.data}")

# Fun√ß√£o para conectar ao banco de dados MySQL
def conectar_banco_Arthur():
    conexao = mysql.connector.connect(
        host="localhost",
        user="root",         # Substitua pelo seu usu√°rio
        password="sua_senha",# Substitua pela sua senha
        database="consultas_arthur"
    )
    return conexao

# Fun√ß√£o para consultar registros
def consultar_banco_Arthur(conexao):
    cursor = conexao.cursor()
    cursor.execute("SELECT * FROM consultas")
    resultados = cursor.fetchall()
    print("\nConsultas registradas no banco de dados:")
    for linha in resultados:
        print(linha)

# Fun√ß√£o para deletar um registro
def deletar_consulta_Arthur(conexao, paciente):
    cursor = conexao.cursor()
    sql = "DELETE FROM consultas WHERE paciente = %s"
    cursor.execute(sql, (paciente,))
    conexao.commit()
    print(f"Consulta de {paciente} deletada com sucesso.")

# =======================
# Testando todas as fun√ß√µes
# =======================

# Conectando ao banco
conexao = conectar_banco_Arthur()

# Criando objetos
consulta1 = Consulta_Arthur("Ana Clara", "Dr. Pedro", "2025-09-15")
consulta2 = ConsultaEspecializada_Arthur("Carlos Eduardo", "Dra. Sofia", "2025-10-20", "Ortopedia")

# Exibindo e verificando
consulta1.exibir_dados_Arthur()
consulta1.verificar_data_Arthur()

consulta2.exibir_dados_Arthur()

# Salvando no banco
consulta1.salvar_no_banco_Arthur(conexao)
consulta2.salvar_no_banco_Arthur(conexao)

# Consultando registros
consultar_banco_Arthur(conexao)

# Deletando um registro
deletar_consulta_Arthur(conexao, "Ana Clara")

# Consultando ap√≥s deletar
consultar_banco_Arthur(conexao)

# Fechando conex√£o
conexao.close()
```
